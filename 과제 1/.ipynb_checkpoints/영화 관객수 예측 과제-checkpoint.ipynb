### 1) 새로운 변수 추가
- 개봉일 변수 안에서 개봉연도와, 개봉 월을 따로 추출하여 새로운 열로 추가
- 또한, 월을 기준으로 계절 열 새롭게 추가

```python
# 계절 열 추가 
def seasonal(x):
    if x == 3 or x == 4 or x == 5:
        return '봄'
    elif x == 6 or x == 7 or x == 8:
        return '여름'
    elif x == 9 or x == 10 or x == 11:
        return '가을'
    else:
        return '겨울'
```

```python
# 문자열을 datatime 타입으로 변경.
def date_prerocessing(data):
    data['release_time'] = data['release_time'].apply(pd.to_datetime)

    data['release_year'] = data['release_time'].apply(lambda x: x.year)
    data['release_month'] = data['release_time'].apply(lambda x: x.month)
    data['release_seasonal'] = data['release_month'].apply(lambda x: seasonal(x))

    # 기존 release_time 변수 삭제
    data.drop('release_time', axis=1, inplace=True)
    
    return data 
```

```python
train_data = date_prerocessing(train_data)

test_data = date_prerocessing(test_data)
```

## 7. 모델링

### 1) 선형 회귀 모델 구현

```python
# 성능 지표를 출력하는 함수 구현


def rmse(y, pred):
    return np.sqrt(mean_squared_error(y, pred))
    
    
def r_2(y, pred):
    return r2_score(y, pred)
    
    
def evaluate_regr(y, pred):
    rmse_val = rmse(y, pred)
    mae_val = mean_absolute_error(y, pred)
    r_score = r_2(y, pred)
    print('R2: {0:.3f}, RMSE: {1:.3f}, MAE: {2:.3f}'.format(r_score, rmse_val, mae_val))
```

```python
# 모델과 학습/테스트 데이터 세트를 입력하면 성능, 평가 수치를 반환 
def get_model_predict(model, X_data, y_target):
    model.fit(X_data, y_target)
    pred = model.predict(X_data)
    
#     로그변환일 경우 다시 기존 값으로 변환
#     y_target = np.expm1(y_target)
#     pred = np.expm1(pred)
    
    print('###', model.__class__.__name__, '###')
    evaluate_regr(y_target, pred)
    print()
```

- 표준화 or 정규화 or 로그변환 안할시 train_dataset 성능 결과

- 표준화 했을시 train_dataset 성능 결과

- 정규화 했을시 train_dataset 성능 결과

- 로그변환 했을시 train_dataset 성능 결과

### 2) 교차검증

```python
def cross_val(model, X_data, y_target, cv):
    # cross_val_score()로 5폴드 세트로 MSE를 구한 뒤 이를 기반으로 다시 RMSE 구함.
    neg_mse_scores = cross_val_score(model, X_data, y_target, scoring="neg_mean_squared_error", cv=cv)
    rmse_scores = np.sqrt(-1 * neg_mse_scores)
    avg_rmse = np.mean(rmse_scores)
    # cross_val_scroe(scoring='neg_mean_squared_error')로 반환된 값은 모두 음수 
    print('###',model.__class__.__name__, '교차검증 결과 ### ')
    print(' 5 folds 의 개별 Negative MSE scores: ', np.round(neg_mse_scores, 2))
    print(' 5 folds 의 개별 RMSE scores: ', np.round(rmse_scores, 2))
    print(' 5 folds 의 평균 RMSE : {0:.3f}'.format(avg_rmse) )
    print()
```

```python
for model in models:
    cross_val(model, X_data, y_target, 5)
```

- 규제모형(릿지, 라쏘)과 앙상블 모델들 모두 하이퍼파라미터가 존재
- 그리드 서치를 통한 하이퍼파라미터 튜닝 과정이 필요하지만 생략
