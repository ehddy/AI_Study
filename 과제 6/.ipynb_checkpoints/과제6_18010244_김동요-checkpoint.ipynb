## 2. 데이터 전처리

### 1) 변수 설정

```python
# 종속 변수
targets = train_data.label.values

# 독립변수 
X_features = train_data.loc[:, train_data.columns != 'label'].values/255 # 정규화
```

-  이미지 데이터의 경우 픽셀 정보를 0과 255 사이의 값으로 가지는데, 이를 255로 나눠줘서 0.0과1.0 사이의 값을 가지게 만들어준다.

### 2) 데이터 분할

```python
# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X_features,
                                                    targets,
                                                    test_size = 0.2,
                                                    random_state = 18010244) 
```

- train 데이터 안에서, 검증을 진행하기 위해, 데이터를 학습용 데이터와 검증용 데이터로 8:2로 분할해주었다. 

### 3) 텐서 변환 및   TensorDataset, DataLoader

- 그래디언트를 파이토치 라이브러리를 통해 구하기 위해 array인 값들을 텐서로 변경해주었다.

```python
X_train = torch.from_numpy(X_train)
X_test = torch.from_numpy(X_test)

y_train = torch.from_numpy(y_train).type(torch.LongTensor)
y_test = torch.from_numpy(y_test).type(torch.LongTensor)
```

-  Pytorch에서 입력 데이터를 쉽게 처리하고, 배치 단위로 잘러서 학습할 수 있게 도와주는 TensorDataset과 DataLoader 모듈을 사용하였다. 

```python
# Pytorch 트레인셋 테스트셋
train = torch.utils.data.TensorDataset(X_train, y_train)

test = torch.utils.data.TensorDataset(X_test, y_test)
```

```python
# 배치 사이즈와 에포크, iteration 설정

batch_size = 100
n_iters = 10000

num_epochs = n_iters / (len(X_train) / batch_size)

num_epochs = int(num_epochs)

# data loader를 통해 배치크기나 데이터를 섞을지 여부(shuffle) 등을 결정
train_loader = DataLoader(train, batch_size = batch_size, shuffle = False)
test_loader = DataLoader(test, batch_size = batch_size, shuffle = False)
```

- 배치크기는 100으로 설정하였으며 이터레이션은 10000으로 설정하여 epoch 수치를 변수로 넣었다.

## 3. 모델 생성

### 1) 로지스틱 회귀 모델

```python
# 로지스틱 함수 구현 
class LogisticRegressionModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LogisticRegressionModel, self).__init__()
        
        # 회귀 모형 생성
        self.linear = nn.Linear(input_dim, output_dim)

    def forward(self, x):
        out = self.linear(x)
        return out
```

```python
# 배치 사이즈와 에포크, iteration 설정

batch_size = 100
n_iters = 10000

num_epochs = n_iters / (len(X_train) / batch_size)

num_epochs = int(num_epochs)
```

- 입력 변수의 차원과, 출력변수의 차원을 지정해주었다. 

```python
# 로지스틱 모델 생성
model = LogisticRegressionModel(input_dim, output_dim)
```

- 손실함수로는 크로스 엔트로피를, 옵티마이저로 SGD(모멘텀, 학습률 0.001)를 선택하였다.

```python
# 크로스 엔트로피 (소프트 맥스도 포함되어 있음)
error = nn.CrossEntropyLoss()

# SGD Optimizer 
learning_rate = 0.001
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
# training 모델을 학습 진행 

count = 0 
loss_list = []
iteration_list = []


for epoch in range(num_epochs):
    for i, (images, labels) in enumerate(train_loader):
        
        # 자동 미분을 해주기 위해 tensor를 Variable로 감싸준다. 
        # 1차원으로 구성된 텐서의 차원을 3차원(1, 28 , 28)으로 변경
        train = Variable(images.view(-1, 28*28))
            
        label = Variable(labels)
        
        # 그래디언트 초기화 
        optimizer.zero_grad()
        
        # 피드포워드 연산 
        outputs = model(train)
        
        # 손실 계산 
        loss = error(outputs, labels)
        
        # 이분 진행 
        loss.backward()
        
        # 가중치 업데이트 
        optimizer.step()
        
        count += 1
        
        # 검증 데이터 평가 
        if count % 50 == 0:
                 
            correct = 0
            total = 0
            
            # iteration이 50일때마다 검증데이터로 평가를 진행 
            for images, labels in test_loader:
                test = Variable(images.view(-1, 28*28))
                
                
                outputs = model(test)
                
                # 예측한 값 중 최대값 
                predicted = torch.max(outputs.data, 1)[1]
                
                total += len(label)
                
                correct += (predicted == labels).sum()
            
            accuracy = 100 * correct / float(total)
            
            
            # 손실 기록 
            loss_list.append(loss.data)
            iteration_list.append(count)
            # iteration이 500일때마다 손실과 검증 데이터로 도출한 정확도 출력
            if count % 500 == 0:
                print('Iteration: {}  Loss: {}  Accuracy: {}%'.format(count, loss.data, accuracy))
```

- Iteration을 9500로 설정했을시 평가 데이터의 정확도가 85.75%가 나타난다. 

```python
# iteration별 손실 시각화 
plt.plot(iteration_list,loss_list)
plt.xlabel("Number of iteration")
plt.ylabel("Loss")
plt.title("Logistic Regression: Loss vs Number of iteration")
plt.show()
```

### 2) 인공신경망 모델

로지스틱 회귀는 복잡성(비선형)이 증가하면 모델의 정확도가 떨어지는 문제가 있기 때문에, 인공신경망 모델을 통해 은닉층의 개수를 늘려서 복잡성이 높은 모델을 추가로 구현해보았다.

```python
class ANNModel(nn.Module):
    
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(ANNModel, self).__init__()
        
        # 784개의 입력 차원을  100개의 은닉층 차원으로 변환 
        self.fc1 = nn.Linear(input_dim, hidden_dim) 
        
        # 은닉층의 활성화 함수로 Relu를 사용 
        self.relu1 = nn.ReLU()
        
        # 또 하나의 은닉층으로 연결하여 100개의 은닉층 노드의 값으로 변환 
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        
        # 2번쨰 은닉층의 활성화 함수로 hyperbolic tangent를 사용 
        self.tanh2 = nn.Tanh()
        
        # 마지막 은닉층의 노드의 차원을 100개로 설정 
        self.fc3 = nn.Linear(hidden_dim, hidden_dim)
        
        # 마지막 은닉층의 활성화 함수로 ELU 사용 
        self.elu3 = nn.ELU()
        
        # 출력 노드의 차원을 10(라벨의 개수)
        self.fc4 = nn.Linear(hidden_dim, output_dim)  
    
    # 피드포워드 연산 
    def forward(self, x):
        
        out = self.fc1(x)

        # 활성화 함수 1
        out = self.relu1(out)
        
        out = self.fc2(out)
        
        # 활성화 함수 2
        out = self.tanh2(out)

        out = self.fc3(out)
        
        # 활성화 함수 3
        out = self.elu3(out)
        
        # 출력층의 활성화 함수 
        out = self.fc4(out)
        return out
```

```python
input_dim = 28*28
hidden_dim = 100
output_dim = 10

# 인공신경망 모델 생성 
model = ANNModel(input_dim, hidden_dim, output_dim)

# 크로스 엔트로피
error = nn.CrossEntropyLoss()

# SGD 
learning_rate = 0.02
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

- 입력층의 차원과, 은닉층의 차원, 그리고 출력층의 차원을 지정해주었으며, 손실함수와 옵티마이져는 이전과 동일하게 설정하였다. 

```python
# 학습데이터 학습(이전과 동일한 방식으로 코딩)
count = 0
loss_list = []
iteration_list = []
accuracy_list = []
for epoch in range(num_epochs):
    for i, (images, labels) in enumerate(train_loader):

        train = Variable(images.view(-1, 28*28))
        labels = Variable(labels)
        
        optimizer.zero_grad()
        
        outputs = model(train)
    
        loss = error(outputs, labels)
        
        loss.backward()
        
        optimizer.step()
        
        count += 1
        
        if count % 50 == 0: 
            correct = 0
            total = 0

            for images, labels in test_loader:

                test = Variable(images.view(-1, 28*28))
                

                outputs = model(test)
                
                predicted = torch.max(outputs.data, 1)[1]
                
                total += len(labels)

                correct += (predicted == labels).sum()
            
            accuracy = 100 * correct / float(total)
            
            loss_list.append(loss.data)
            iteration_list.append(count)
            accuracy_list.append(accuracy)
        if count % 500 == 0:
            print('Iteration: {}  Loss: {}  Accuracy: {} %'.format(count, loss.data, accuracy))
```

- Iteration을 9500로 설정했을시 평가 데이터의 정확도가 96.26% 나타나며, 로지스틱 모델보다, 성능이 훨씬 높게 나타난다.

```python
# iteration별 손실 시각화 
plt.plot(iteration_list,loss_list)
plt.xlabel("Number of iteration")
plt.ylabel("Loss")
plt.title("ANN: Loss vs Number of iteration")
plt.show()

# iteration별 정확도 시각화 
plt.plot(iteration_list,accuracy_list,color = "red")
plt.xlabel("Number of iteration")
plt.ylabel("Accuracy")
plt.title("ANN: Accuracy vs Number of iteration")
plt.show()
```

- Iteration이 높을수록 손실이 감소하며, 정확도도 늘어나는 추세를 보여준다.

### 3) 합성곱 신경망(CNN)

- 이미지 처리에 탁월한 성능을 보이는 합성곱 신경망을 마지막으로 구현해보았다. 

```python
# 합성곱 신경망 클래스 생성 
class CNNModel(nn.Module):
    def __init__(self):
        super(CNNModel, self).__init__()
        
        
        # 첫 번째 합성곱 층 
        # 입력 채널이 1이고, 출력채널을 16, 커널 사이즈 5, 스트라이드 1로 지정 
        self.cnn1 = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=5, stride=1, padding=0)
        # 활성화 함수를 통과한 다음에, 
        self.relu1 = nn.ReLU()
        
        # 2사이즈의 커널로 맥스 풀링을 진행 
        self.maxpool1 = nn.MaxPool2d(kernel_size=2)
     
        # 두 번째 합성곱 층 
        # 입력 채널이 16(이전 층의 출력 채널)이고, 출력채널을 32, 커널 사이즈 5, 스트라이드 1로 지정 
        self.cnn2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=5, stride=1, padding=0)
        # 활성화 함수를 통과한 다음에, 
        self.relu2 = nn.ReLU()
        
        # 2사이즈의 커널로 맥스 풀링을 진행 
        self.maxpool2 = nn.MaxPool2d(kernel_size=2)
        
        # 전결합층 4x4x32 입력하여 10개의 아웃풋이 나오게 설정 
        self.fc1 = nn.Linear(32 * 4 * 4, 10) 
    
    def forward(self, x):
        # 첫 번째 합성곱 층 
        out = self.cnn1(x)
        out = self.relu1(out)
        

        out = self.maxpool1(out)
        
        # 두 번째 합성곱 층 
        out = self.cnn2(out)
        out = self.relu2(out)
     
        out = self.maxpool2(out)
        
        # 전결합층을 위해서 Flatten
        out = out.view(out.size(0), -1)

        # 전결합 연산 
        out = self.fc1(out)
        
        return out
```

```python
# 배치 사이즈와 에포크, iteration 설정

batch_size = 100
n_iters = 2500

num_epochs = n_iters / (len(X_train) / batch_size)

num_epochs = int(num_epochs)
```

- 배치크기는 100으로 설정하였으며 이터레이션은 전보다 낮은 2500으로 설정하여 epoch 수치를 변수로 넣었다.

```python
# CNN 모형 생성 
model = CNNModel()

# 크로스엔트로피 손실함수(소프트맥스 포함)
error = nn.CrossEntropyLoss()

# 옵티마이저 SGD(이전과 동일)
learning_rate = 0.1
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
# CNN 모델 학습(이전과 코딩 방식 동일)
count = 0
loss_list = []
iteration_list = []
accuracy_list = []
for epoch in range(num_epochs):
    for i, (images, labels) in enumerate(train_loader):
        
        # 입력층의 차원을 이미지 형식에 맞게 변환 
        train = Variable(images.view(100,1,28,28))
        labels = Variable(labels)
        
        optimizer.zero_grad()

        outputs = model(train)
        
        loss = error(outputs, labels)
        
        loss.backward()
        
        optimizer.step()
        
        count += 1
        
        if count % 50 == 0:      
            correct = 0
            total = 0
            for images, labels in test_loader:
                
                test = Variable(images.view(100,1,28,28))
                
                outputs = model(test)
                
                predicted = torch.max(outputs.data, 1)[1]
                
                total += len(labels)
                
                correct += (predicted == labels).sum()
            
            accuracy = 100 * correct / float(total)

            loss_list.append(loss.data)
            iteration_list.append(count)
            accuracy_list.append(accuracy)
        if count % 500 == 0:
            # Print Loss
            print('Iteration: {}  Loss: {}  Accuracy: {} %'.format(count, loss.data, accuracy))
```

- Iteration을 2000로 설정했을시 평가 데이터의 정확도가 98.09% 나타나며, 이전의 모델들보다 성능이 뛰어나다.

```python
# iteration별 손실 시각화 
plt.plot(iteration_list,loss_list)
plt.xlabel("Number of iteration")
plt.ylabel("Loss")
plt.title("CNN: Loss vs Number of iteration")
plt.show()

# iteration별 정확도 시각화 
plt.plot(iteration_list,accuracy_list,color = "red")
plt.xlabel("Number of iteration")
plt.ylabel("Accuracy")
plt.title("CNN: Accuracy vs Number of iteration")
plt.show()
```

- Iteration이 높을수록 손실이 감소하며, 정확도도 늘어나는 추세를 보여준다. 인공신경망의 모델보다, 손실과, 정확도의 편차가 크게 나타난다.
